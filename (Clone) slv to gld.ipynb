```python
from pyspark.sql import functions as F
```

```python
df_silver = spark.table("slv.sales")
display(df_silver)
```

```python
spark.sql("CREATE SCHEMA IF NOT EXISTS gld")
```

```python
dim_product = df_silver.select(
    "ProductID",
    "Product",
    "Category",
    "Segment",
    "ManufacturerID",
    "Manufacturer"
).dropDuplicates()

display(dim_product)

dim_product.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("gld.dim_product")
```

```python
dim_customer = df_silver.select(
    "CustomerID",
    "email",
    "first_name",
    "last_name",
    "City",
    "State",
    "Region",
    "ZipCode",
    "Country"
).dropDuplicates()

display(dim_customer)

dim_customer.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("gld.dim_customer")
```

```python
dim_campaign = df_silver.select(
    "CampaignID"
).dropDuplicates()

display(dim_campaign)

dim_campaign.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("gld.dim_campaign")
```

```python
fact_sales = df_silver.select(
    "Date",
    "CustomerID",
    "ProductID",
    "CampaignID",
    "Units",
    F.col("Unit_Cost").alias("Unit_Cost"),
    F.col("Unit_Price").alias("Unit_Price")
)

display(fact_sales)

fact_sales.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("gld.fact_sales")
```
